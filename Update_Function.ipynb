```python
import psycopg2
import sys

# Function to encrypt data using Caesar cipher
def encrypt_caesar(data, key):
    encrypted_data = ''
    for char in data:
        if char.isalpha():
            shift = key % 26  # Ensure the shift is within the alphabet range
            ascii_offset = 65 if char.isupper() else 97  # ASCII offset for uppercase or lowercase
            encrypted_char = chr((ord(char) - ascii_offset + shift) % 26 + ascii_offset)
        else:
            # For non-alphabetic characters, apply the same shift as the key
            encrypted_char = chr((ord(char) + key) % 256)
        encrypted_data += encrypted_char
    return encrypted_data

# # Function to select data from the "Client_Credentials" table
# def select_from_table(db_params):
#     try:
#         # Establish a connection to the PostgreSQL database
#         conn = psycopg2.connect(**db_params)
#
#         # Create a cursor to execute queries
#         cursor = conn.cursor()
#
#         query = """SELECT * FROM public.dim_client_credentials"""
#
#         # Execute the query
#         cursor.execute(query)
#         rows = cursor.fetchall()
#         return rows
#
#     except psycopg2.Error as e:
#         print("Error: ", e)
#
#     finally:
#         # Don't forget to close the cursor and connection when you're done.
#         if 'cursor' in locals():
#             cursor.close()
#         if 'conn' in locals():
#             conn.close()

# Function to update data in the "Client_Credentials" table with encrypted values
def update_table_with_encrypted_data(db_params,clientId,Name,password,email):
    try:
        # Establish a connection to the PostgreSQL database
        conn = psycopg2.connect(**db_params)

        # Create a cursor to execute queries
        cursor = conn.cursor()
        #
        # rows = select_from_table(db_params)
        # for i in rows:
        enc_clientname = encrypt_caesar(Name, 7)
        enc_password = encrypt_caesar(password, 7)
        enc_email = encrypt_caesar(email, 7)

        query = '''UPDATE public.dim_client_credentials
                       SET 
                           encrypted_clientname = %s,
                           encrypted_password = %s,
                           encrypted_email = %s
                       WHERE client_id = %s'''

        # Execute the query with the encrypted data and timestamp
        cursor.execute(query, (enc_clientname, enc_password, enc_email, clientId))
        conn.commit()

    except psycopg2.Error as e:
        print("Error: ", e)

    finally:
        # Don't forget to close the cursor and connection when you're done.
        if 'cursor' in locals():
            cursor.close()
        if 'conn' in locals():
            conn.close()

# Database connection parameters
db_params = {
    'host': 'localhost',         # Replace with the actual server address
    'database': 'clientdb',       # Replace with your actual database name
    'user': 'postgres',          # Replace with your PostgreSQL username
    'password': 'password'          # Replace with your PostgreSQL password
}

print('Enter Client ID to be updated :')
clientId = sys.argv[1]          ##to enter value via command line
#clientId=input()               ## for local testing
print('Enter Client Name :')
#Name=input()
Name = sys.argv[2]
print('Enter password :')
#password=input()
password = sys.argv[3]
print('Enter email :')
#email=input()
email = sys.argv[4]
# Run the update function to encrypt and update the data in the table
update_table_with_encrypted_data(db_params,clientId,Name,password,email)

print("Data encryption and update completed.")
```
